# EDA


```python
# EDA of sex, alcohol, stress, walk, gym

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
# sex
sex = ['여자', '남자']
# alcohol
alcohol = ['안마심','1년간 안마심','월 1회미만','월 1회','월 2~4회','주 2~3회','주 4회 이상']

# stress
stress = ['조금', '많이','거의x','대단히']

# walk
walk = ['매일','전혀','주5','주3','주2','주4','주1','주6']

# gym
gym = ['전혀','주2','주3','5일이상','주1','주4']

# metabolic
metabolic = ['No','Yes']

dataset_list = ['sex','alcohol','stress','walk','gym','metabolic']
label_list = [sex,alcohol,stress,walk,gym,metabolic]

fig = plt.figure()
for num in range(len(dataset_list)):
    i = num//3
    j = num%3
    ax = fig.add_subplot(3, 3, num+1)
    fig.set_figheight(20)
    fig.set_figwidth(20)
    ax.pie(dataset[dataset_list[num]].value_counts(),labels=label_list[num],autopct='%.1f%%',startangle=260,counterclock=False, 
                           shadow = True, colors = ['lightblue','lightgreen','#ff9999','#ffc000','#8fd9b6','#d395d0','lavender', 'steelblue'])
                           
    ax.set_title(dataset_list[num])

plt.subplots_adjust(wspace=1)
plt.show()
```

```python
# smoke 수
fig1 = sns.countplot(x="smoke",data=dataset)
fig1.figure.set_size_inches(20,10)
# print(dataset["smoke"].value_counts())
```

```python
# age 분포
sns.set_palette("Dark2")
fig1 = sns.displot(x="age",data=dataset)
fig1.figure.set_size_inches(20,10)   
# sns.set(rc={"figure.figsize":(30, 20)}) 
# print(dataset["age"].value_counts())
```

```python
# time wake of weekdays
sns.set_palette("Dark2")
fig1 = sns.displot(x="time_w", data=dataset)
fig1.figure.set_size_inches(10,5)
# print(dataset["time_w"].value_counts())

# time wake of weekend
sns.set_palette("Dark2")
fig1= sns.displot(x="time_wk", data=dataset)
fig1.figure.set_size_inches(10,5)
# print(dataset["time_w"].value_counts())
```
